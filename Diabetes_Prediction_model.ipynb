1. **Problem Definition**
* predicting whether a person has diabetes based on medical features.

2. **Project Motivation & Background**
* Diabetes is a growing global health challenge that can lead to serious complications if not detected early.
* Many individuals, especially in developing countries, face delayed diagnosis due to limited access to timely and affordable medical screening. Early identification of high-risk individuals is therefore essential for effective disease management and prevention.
* This project is motivated by the need to use machine learning to support early diabetes detection by analyzing readily available health data.
* By building a supervised classification model in Python, the project demonstrates how data-driven techniques can assist healthcare decision-making and promote preventive care, while complementing—not replacing—medical diagnosis.

3. Define the Machine Learning Task
* The objective is to develop a model that learns from historical, labeled medical data to predict whether an individual has diabetes or not.

* The input to the model consists of multiple health-related features such as glucose level, body mass index (BMI), age, blood pressure, insulin level, and other clinical measurements.

8. Split the Data into Train and Test Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

9. Training the model

```python
Classifier = svm.SVC(kernel='linear')
```

```python
#training the Support Vector Machine Classifier
Classifier.fit(X_train, Y_train)
```

10. Model Evaluation

a) Accuracy score

```python
# Accuracy score on the traing data
X_train_prediction = Classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy score of the training data : ', training_data_accuracy)
```
